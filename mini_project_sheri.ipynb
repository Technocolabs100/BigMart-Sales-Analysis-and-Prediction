## Encoding

```python
columns_to_encode = ['Item_Identifier', 'Item_Fat_Content','Item_Type','Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

for column in columns_to_encode:
    le = LabelEncoder()
    updated2_df[column] = le.fit_transform(updated2_df[column])

print(updated2_df.head())
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score
```

```python
updated2_df.describe()
```

```python
updated2_df.info()
```

```python
X = updated2_df[['Item_Identifier', 'Item_Fat_Content','Item_Visibility','Item_Type','Item_MRP','Outlet_Identifier','Outlet_Establishment_Year','Outlet_Location_Type','Outlet_Location_Type','Outlet_Type']]
y = updated2_df['Item_Outlet_Sales'].values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print('Root Mean Squared Error:', rmse)
print('R-squared Score:', r2)
```

```python
train_score = model.score(X_train, y_train)
print('Training Score:', train_score)
```

```python
test_score = model.score(X_test, y_test)
print('Testing Score:', test_score)
```
